# Constants
- Fixed values(strings, numbers, letters, etc.) and variable names can be counted as constants
- Also Reserved words(which cannot be used as variable names) are constants


# Variable Naming Rules
- it must start with a letter or underscore
- cannot consist characters other than letters, numbers or some exceptional symbols
- python is a case sensitive language(which means you cannot change any character of the variable name)


```python
# valid variable namings but not mnemonic
a = 2
__a = 2
a1 = 2
```

```python
# those two variables are different because python is a case sensitive language
b = 1
B = 2
print(B)
print(b)
```

```python
123 = 3 # not a valid variable naming
```

# Mnemonic
- a naming method which explains logic of the code to help the developer who is looking to the code

```python
# mnemonic way to write code
hour = 5
rate = 6
print(hour * rate)
```

# Assignment Statements
### Assigning to a variable:

```python
x = 4 + 4
print(x)
```

- this shows when the process of the right side is done then it stores the result on left side to let it be allocated in the memory 


### Reassigning to a variable:

```python
x = 6 + 6
print(x)
```

- if x was assigned before you still can reassign it again with a similar process to have a new output which also overrides old value

# Operator precedence:

```python
print((3+5)*3**2/3//4 + 3 - 2)
```
